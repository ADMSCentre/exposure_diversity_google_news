## Closing Remarks

Upon observation, the following remarks can be made about the participants of the Australian Search Experience, in comparison with data obtained from the Australian Bureau of Statistics:

* The participants are predominantly "Male", although non-binary gender types are comparable.
* There is a good representation of the majority of states of residence, however given the size of the participant sample, its expected that there may not be enough corresponding data donations for individuals residing in the Northern Territory, the Australian Capital Territory, or Tasmania in order to report adequately about these responses, as will become evident in consequent analyses.
* There is an under-representation of retirees among the participants, whereas other responses for participants' employment statuses are over-represented.
* Pasrticipants aged 35 - 64 are over-represented, while participants aged 25 - 34 are under-represented.
* The vast majority of participants have completed tertiary education, while other responses are under-represented.
* Participants earning $104,000 - $155,999, as well as participants with 'Undescribed' income brackets are under-represented, while participants in other income brackets are over-represented.
* Among political party preferences, participants who prefer the Liberal National Party are under-represented, while participants who prefer the Australian Greens are over-represented.
