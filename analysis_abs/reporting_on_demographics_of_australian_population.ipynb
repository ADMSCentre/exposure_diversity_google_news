## Interpretation Of Datasets

The process by which each of these data sources was interpreted are contained in the `census.py` software implementation, for which the interpretation is comprehensively explained in __RQ1 - Representing Australia__ section of the accompanying technical report __Technical Report - Establishing Statistical Significance In ADM+S Data Donation Projects__. Both of these files are contained within this repository.

_Note: While the technical report is dated to the 2016 Australian Census, the methodology is near-identically adapted to the 2021 Australian Census, the only notable difference being the contained table numbers "G43" and "G46" of the 2016 Australian Census datapacks changing to "G46" and "G49" respectively in the 2021 Australian Census datapacks._

Upon comprehension of all datasets, a joint distribution of the following relevant demographic characteristics is given below:

* Age
* Gender
* State Of Residence
* Employment Status
* Education Level
* Income Bracket
* Political Party Preference

To assist in the summarisation and comparison of the Australian Census data, certain responses attributed to these demographic characteristics have been aggregated - again, this is detailed in the aforementioned technical report.
