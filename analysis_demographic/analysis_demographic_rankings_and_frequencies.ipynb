## Influence Of Demographics Upon News Source Prominence

For each of the keywords, the search results returned by Google News reflect prominent news sources - prominence of a news source can be established by how frequently it appears, or where it would (on average) be ranked within the total list of search results.

Here we examine whether individuals pertaining to certain demographic characteristics observe significant changes in the prominence of news sources, as opposed to others.

### Normality Tests

It is possible for an observed effect (of a change in prominence of a news source as a result of some demographic characteristic's response) to be the result of a false reading.

One such instance in which this can occur is when there is not enough observations of a news source to reach a significant conclusion - this would introduce high variance within the sample data, producing erratic results that could not be deemed significant.

An example can be seen in the graph below for the news source __Blue Mountains Gazette__, returned for the queried keyword __'Anthony Albanese'__ on Google News, and segmented by participants' age. Individuals who did not report an age (as well as individuals over 75) in this case presented non-uniform search result rankings compared to those who did. 

```python
compact_plot_average_search_rankings_vs_frequencies("k1", "Anthony Albanese", "age", "bluemountainsgazette.com.au (AU)", "frq")
```

In our tests, we allow for an error margin of 5%, which permits p-values that are less than or equal to 0.05.

We are also mindful of instances of where normality tests may not be as helpful, such as in cases where there are a significant number of observations, however the results remain highly varied. We report on normality tests statistics (in the attached `contextualised_set_summarised_by_domain.json` file), and disregard instances where observed effects are unsubstantially distinct.

We should expect that for demographic characteristics to have influence over the news sources, that there should be observably substantial differences between the responses obtained for the various demographic characteristics being visualised.

### Confounding Effect Of Frequencies On Representation

A confounding effect that has contaminated this aspect of the analysis has been the issue of fewer observations for a news source naturally yielding fewer participant observations from particular demographics. This can adversely distort representation in certain cases.

For instance, the majority of participants who reported observations within the research project were found to predominantly reside in New South Wales. This would mean that both locations were adversely overrepresented, confounding naturally emerging effects that might be present in the data.

Specifically, consider for the news source __The West Australian__ observed on the Google News 'COVID' keyword, the following participant responses distributed by state of residence.

```python
plot_representation_distributions("k1", "Barnaby Joyce", "state", "thewest.com.au (AU)", show=True)
```

While the geographical targeting towards residents of Queensland should be expected, this is not reported above, and is most likely due to overrepresentation of constituents from New South Wales, which consequently distorts the result.
